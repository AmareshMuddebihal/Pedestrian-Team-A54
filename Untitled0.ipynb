```python
import os
import shutil

# Define the source directory where your files are located
source_directory = "/content/drive/MyDrive/capstone/vehicles (1)/vehicles"

# Define the destination directories for JPG and TXT files
jpg_destination = "/content/drive/MyDrive/capstone/all_img"
txt_destination = "/content/drive/MyDrive/capstone/all_label"

# Ensure the destination directories exist; create them if necessary
os.makedirs(jpg_destination, exist_ok=True)
os.makedirs(txt_destination, exist_ok=True)

# List files in the source directory
for filename in os.listdir(source_directory):
    # Get the file extension
    file_extension = filename.split(".")[-1].lower()

    # Create the source file path
    source_file = os.path.join(source_directory, filename)

    # Check if it's a JPG file and move it to the JPG destination
    if file_extension == "jpg":
        shutil.move(source_file, os.path.join(jpg_destination, filename))

    # Check if it's a TXT file and move it to the TXT destination
    elif file_extension == "txt":
        shutil.move(source_file, os.path.join(txt_destination, filename))

# Optional: Print a message to confirm the separation
print("Files separated successfully!")
```
