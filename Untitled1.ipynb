```python
import zipfile
import os

def merge_zip_contents(zip_file1, zip_file2, output_folder):
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Extract contents of the first zip file
    with zipfile.ZipFile(zip_file1, 'r') as zip1:
        zip1.extractall(output_folder)

    # Extract contents of the second zip file and append to the output folder
    with zipfile.ZipFile(zip_file2, 'r') as zip2:
        for file_info in zip2.infolist():
            # Check if the file already exists in the output folder
            while os.path.exists(os.path.join(output_folder, file_info.filename)):
                # If yes, modify the filename to avoid overwriting
                file_info.filename = "_" + file_info.filename

            os.path.join(output_folder, file_info.filename)
if __name__ == "__main__":
    zip_file1 = "/content/drive/MyDrive/all_img.zip"
    zip_file2 = "/content/drive/MyDrive/all_label.zip"
    output_folder = "/content/drive/MyDrive/capstone/all_img_label"

    merge_zip_contents(zip_file1, zip_file2, output_folder)
```
